```python
# Load each dataset with only selected features
df1 = pd.read_csv("Database of rock burst samples.csv")[['NO.', '??(MPa)', '?c(MPa)', '?t(MPa)', '??/?c', '?c/?t', 'Wet']]
df1 = df1.rename(columns={
    '??(MPa)': 'sigma_theta',
    '?c(MPa)': 'sigma_c',
    '?t(MPa)': 'sigma_t',
    '??/?c': 'stress_ratio',
    '?c/?t': 'brittleness_index'
}).dropna()

df2 = pd.read_csv("Rockburst Sample Database.csv")[['H/m', '??/Mpa', '?c/Mpa', '?t/Mpa', '??/?c', '?c/?t', 'Wet', '??????']]
df2 = df2.rename(columns={
    'H/m': 'depth',
    '??/Mpa': 'sigma_theta',
    '?c/Mpa': 'sigma_c',
    '?t/Mpa': 'sigma_t',
    '??/?c': 'stress_ratio',
    '?c/?t': 'brittleness_index',
    '??????': 'rockburst_grade'
}).dropna()

df3 = pd.read_csv("sustainability-2534712-supplementary.csv")[['SCF', 'B1', 'B2', 'D(m)', 'Rockburst grade']]
df3 = df3.rename(columns={
    'SCF': 'stress_ratio',
    'B1': 'brittleness_index',
    'B2': 'rock_quality_index',
    'D(m)': 'depth',
    'Rockburst grade': 'rockburst_grade'
}).dropna()

df4 = pd.read_csv("Rockburst dataset in tunnels.csv")[['D1\n(unit)', 'D2\n(J)', 'D3\n(m3)', 'Level']]
df4 = df4.rename(columns={
    'D1\n(unit)': 'seismic_event_count',
    'D2\n(J)': 'seismic_energy',
    'D3\n(m3)': 'fracture_volume',
    'Level': 'rockburst_level'
}).dropna()

df5 = pd.read_csv("seismic-bumps.csv")[['genergy', 'gpuls', 'ghazard', 'class']]
df5 = df5.rename(columns={
    'genergy': 'total_seismic_energy',
    'gpuls': 'seismic_pulses',
    'ghazard': 'geological_hazard_class',
    'class': 'rockburst_binary'
}).dropna()

# Standardize target variables (example conversion)
df4['rockburst_grade'] = df4['rockburst_level'].map({
    'Slight': 1,
    'Moderate': 2,
    'Intense': 3
})

df5['rockburst_grade'] = df5['rockburst_binary'].map({
    0: 0,  # No rockburst
    1: 3   # Treat binary "1" as severe rockburst
})

# Combine datasets with common features
combined_df = pd.concat([
    df1[['sigma_theta', 'sigma_c', 'sigma_t', 'stress_ratio', 'brittleness_index', 'Wet']],
    df2[['sigma_theta', 'sigma_c', 'sigma_t', 'stress_ratio', 'brittleness_index', 'Wet', 'depth', 'rockburst_grade']],
    df3[['stress_ratio', 'brittleness_index', 'rock_quality_index', 'depth', 'rockburst_grade']],
    df4[['seismic_energy', 'fracture_volume', 'rockburst_grade']],
    df5[['total_seismic_energy', 'seismic_pulses', 'rockburst_grade']]
], axis=0, ignore_index=True)

# Fill missing values if needed (example)
combined_df['depth'] = combined_df['depth'].fillna(combined_df['depth'].median())
combined_df['Wet'] = combined_df['Wet'].fillna(combined_df['Wet'].mean())

# Save final dataset
combined_df.to_csv("combined_rockburst_dataset.csv", index=False)
print("Combined dataset saved with shape:", combined_df.shape)
```

# "rockburst_grade" is the target column

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Prepare data (drop rows where target is missing)
df = combined_df.dropna(subset=['rockburst_grade']).copy()

# Separate features and target
X = df.drop('rockburst_grade', axis=1)
y = df['rockburst_grade']

# Basic imputation (fill missing with median)
X = X.fillna(X.median())

# Train a quick model
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Get importance scores
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False)

print(importance)
```

```python
from sklearn.inspection import permutation_importance

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model.fit(X_train, y_train)

result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)
perm_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': result.importances_mean
}).sort_values('Importance', ascending=False)
```

```python
# For continuous targets
correlation = df.corr()['rockburst_grade'].abs().sort_values(ascending=False)

# For categorical targets
from sklearn.feature_selection import mutual_info_classif
mi_scores = mutual_info_classif(X, y, random_state=42)
mi_df = pd.DataFrame({'Feature': X.columns, 'MI_Score': mi_scores}).sort_values('MI_Score', ascending=False)
```
